```python
#plot sepal length vs. sepal width

import csv
import urllib2
from matplotlib import pylab as plt

url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/bezdekIris.data'
c_file = csv.reader(urllib2.urlopen(url)) #reading CSV file into c_file structure

sepal_length = [] #initialise array for sepal length
sepal_width = [] #initialise array for sepal width
colors = [] #initialise array data point colors

for row in c_file: 
    if not row:
        break #break if an empty line is passed
    else:
        sepal_length.append(row[0]) #creating sepal length array
        sepal_width.append(row[1]) #creating sepal width array
        if row[4] == "Iris-setosa":
            colors.append("r") #defining red data point for Iris Setosa
        elif row[4] == "Iris-versicolor":
            colors.append("y") #defining yellow data point for Iris Versicolor
        elif row[4] == "Iris-virginica":
            colors.append("b") #defining blue data point for Iris Virginica
         
plt . scatter ( sepal_length , sepal_width , c = colors ) #plotting graph
plt . suptitle('Sepal Width vs. Sepal Length', fontsize=20)
plt . xlabel('sepal length', fontsize=16)
plt . ylabel('sepal width', fontsize=16)
plt . savefig ("sepal_width_vs_sepal_length.png") #saving graph as .png file
```
